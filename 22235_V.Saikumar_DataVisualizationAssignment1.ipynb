```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_month = pd.read_csv("accidents_2017.csv")

df_month.shape
df_month.head()
```

# BAR CHARTS

```python
acc_part_of_day = pd.pivot_table(df_month,index= 'Part of the day', values= 'Victims',aggfunc=np.sum)
acc_part_of_day
```

```python
plt.style.use('seaborn-ticks')
plt.figure(figsize=(7,7))
       
plt.bar(acc_part_of_day.index, acc_part_of_day['Victims'], color= "lawngreen")
plt.scatter(acc_part_of_day.index, acc_part_of_day['Victims'], color= "purple")


plt.xlabel('no of victims affected by accidents during which part of the day',fontsize = 15)
plt.ylabel('count',fontsize = 15)
plt.title(' accidents  in barcelona',fontsize = 20)

plt.savefig('3. Bargraph.png', dpi=300)


plt.show()
```

```python
acc_month = pd.pivot_table(df_month,index= 'Month', values= 'Victims',aggfunc=np.sum)
acc_month
```

```python
plt.style.use('seaborn-ticks')
plt.figure(figsize=(10,10))
       
plt.bar(acc_month.index, acc_month['Victims'], color= "orange")
plt.scatter(acc_month.index, acc_month['Victims'], color= "purple")

plt.xticks(rotation=70)
plt.xlabel('month',fontsize = 15)
plt.ylabel('no of victims affected by accidents',fontsize = 15)
plt.title('total accidents in a year in barcelona',fontsize = 20)
plt.ylim(700,1300)
plt.savefig('1. Bargraph.png', dpi=300)


plt.show()
```

```python
acc_day = pd.pivot_table(df_month,index= 'Weekday', values= 'Victims',aggfunc=np.sum)
acc_day
```

```python
acc_day["Victims"].sum()
```

```python
plt.figure(figsize=(10,10))

plt.bar(acc_day.index, acc_day['Victims'],color = "fuchsia",label="weekday")
plt.scatter(acc_day.index, acc_day['Victims'],color="black")
plt.xticks(rotation=70)
plt.xlabel('week days',fontsize = 20)
plt.ylabel('no of victims affected by accidents',fontsize = 20)
plt.title('accidents occured in weekdays in barcelona',fontsize = 20)
plt.ylim(750,2200)
plt.legend()
plt.savefig('2. Bargraph.png', dpi=300)
plt.show()
```

# Pie Charts

```python
acc_part_of_day = pd.pivot_table(df_month,index= 'Part of the day', values= 'Victims',aggfunc=np.sum)
acc_part_of_day
```

```python
plt.figure(figsize = (9,9))
plt.pie(acc_part_of_day.Victims,labels=["Afternoon","Morning","Night"],
       autopct = "%0.3f",
       explode = [0.1,0,0],
       textprops={'fontsize': 14},
       shadow=True)
plt.title("Number of accidents took place in different parts of the day",fontsize = 25)
plt.show()
```

```python
acc_day = pd.pivot_table(df_month,index= 'Weekday', values= 'Victims',aggfunc=np.sum)
acc_day
```

```python
plt.figure(figsize = (9,9))
plt.pie(acc_day.Victims,labels=["Friday","Monday","Saturday","Sunday","Thursday","Tuesday","Wednesday"],
       autopct = "%0.3f",
       explode = [0.2,0,0,0,0,0,0],
       textprops={'fontsize': 14},
       shadow=True)
plt.title("Number of accidents occured in different days of the week",fontsize = 25)

plt.show()
```
