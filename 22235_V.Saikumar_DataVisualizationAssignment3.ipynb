## Line Plot

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_death = pd.read_csv("deaths.csv")
df_death.shape
df_death.head()
```

```python
dist_name = pd.pivot_table(df_death,index="District.Name", values = "Number", aggfunc = np.sum)
dist_name
```

```python
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(15,9))
plt.title("Total number of deaths according to Districts",fontsize=30)
plt.xlabel("District name",fontsize=20)
plt.ylabel("Number of deaths ",fontsize=20)
plt.plot(dist_name.index,dist_name.Number,color="orange")
#plt.scatter(dist_name.index,dist_name.Number,color="black")

#plt.xticks(rotation=45)
```

## Scatter Plot

```python
df_imi_emi = pd.read_csv("immigrants_emigrants_by_age.csv")
df_imi_emi.shape
df_imi_emi.head()
```

```python
immigrants = pd.pivot_table(df_imi_emi,index = "Age",values = "Immigrants" ,aggfunc = np.sum)
immigrants
```

```python
emmigrants = pd.pivot_table(df_imi_emi,index = "Age",values = "Emigrants" ,aggfunc = np.sum)
emmigrants
```

```python
plt.style.use("classic")
plt.figure(figsize=(12,7))
plt.title("Total number of Immigrants and Emigrants according to age",fontsize=20)
plt.xlabel("Age",fontsize=20)
plt.ylabel("No of immigrants and emigrants ",fontsize=20)
plt.xticks(rotation=60)

plt.scatter(immigrants.index,immigrants.Immigrants,color="orange")
plt.scatter(emmigrants.index,emmigrants.Emigrants,color="purple")
plt.legend(["immigrants","emigrants"])
```

## Subplots

```python
plt.style.use("fivethirtyeight")

fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(20,20))

ax[0].plot(immigrants.index,immigrants.Immigrants,color="orange")
ax[0].set_xlabel('Age',fontsize=30) 
ax[0].set_ylabel('immigrants',fontsize=30) 
ax[0].set_title('Number of immigrants',fontsize=30)


ax[1].plot(emmigrants.index,emmigrants.Emigrants,color="purple") 
ax[1].set_xlabel('Age',fontsize=30) 
ax[1].set_ylabel('Emigrants',fontsize=30) 
ax[1].set_title('Number of emigrants',fontsize=30) 
```
