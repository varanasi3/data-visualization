```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df_names = pd.read_csv("most_frequent_names.csv")
df_names.shape
df_names.head()
```

```python
c_names={}
for i in df_names['Name'].unique():
    
    c_names[i]=df_names[df_names['Name']== i].Frequency.unique()
```

```python
plt.figure(figsize=(20,10))
plt.style.use("_classic_test_patch")
plt.boxplot([x for x in c_names.values()], 
           labels = [x for x in c_names.keys()],showmeans ="True")
plt.xlabel('Name',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.ylim(0,5000)
plt.xlim(0,7.5)
plt.title('Some of the popular names in past deacades', fontsize=20)
plt.grid()
plt.savefig("1.BoxPlot.png",dpi=300)
plt.show()
```

```python
df_birth = pd.read_csv("births.csv")
df_birth.shape
df_birth.head()
```

```python
c_birth={}
for i in df_birth['District Name'].unique():
    
    c_birth[i]=df_birth[df_birth['District Name']== i].Number.unique()
```

```python
plt.figure(figsize=(20,10))
plt.style.use("fast")
plt.boxplot([x for x in c_birth.values()], 
           labels = [x for x in c_birth.keys()],showmeans="True")
plt.xlabel('District Name',fontsize=15)
plt.ylabel('Frequency of births in each districts',fontsize=15)
plt.xticks(rotation = 30)
plt.title('Birth rates', fontsize=20)
plt.savefig("2.BoxPlot.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.style.use("seaborn-deep")
plt.hist(df_birth['District Name'],rwidth=0.9, alpha = 0.4, color= 'green', bins = 10 ,histtype="stepfilled")
plt.xlabel('District_Name',fontsize="10")
plt.ylabel('Frequency ofBirths',fontsize="10")
plt.title(' Birth rate in the Districts',fontsize="30")
plt.xticks(rotation = 45)
plt.savefig("3.Histogram.png",dpi=300)
plt.show()
```

```python
df_emi = pd.read_csv("immigrants_emigrants_by_age.csv")
df_emi.shape
df_emi.head()
```

```python
c_immi={}
for i in df_emi['Year'].unique():
    
    c_immi[i]=df_emi[df_emi['Year']== i].Immigrants
c_immi
```

```python
c_emmi={}
for i in df_emi['Year'].unique():
    
    c_emmi[i]=df_emi[df_emi['Year']== i].Emigrants
c_emmi
```

```python
plt.figure(figsize=(10,5))
plt.style.use("tableau-colorblind10")

plt.subplot(1,2,1)
plt.hist(x=c_immi.values(),label=list(c_immi.keys()))
plt.ylabel('Frequency of immigrants ',fontsize="10")
plt.legend()
plt.grid()

plt.subplot(1,2,2)
plt.hist(x=c_emmi.values(),label=list(c_emmi.keys()))
plt.ylabel('Frequency of  emmigrants',fontsize="10")
plt.legend()
plt.grid()


plt.title('Emigrants and Immigrants  in years',fontsize="30")


plt.tight_layout()
plt.savefig("3.Histogram.png",dpi=300)
plt.show()
```
